# 1.   Crawler inicial a partir da Wikipédia

Neste exercício, foi desenvolvido um crawler simples a partir da página da Wikipédia sobre Thomas Bayes.

O objetivo é:

- baixar o HTML da página inicial;
- identificar os links internos da página;
- considerar apenas os links cujo `href` começa com `/wiki/`;
- montar as URLs completas desses links;
- ao final baixar os HTMLs completos das páginas encontradas.

Para isso, o programa utiliza requisição HTTP com `requests` e análise do HTML com `BeautifulSoup`.

## A. Importação das bibliotecas

Nesta etapa, são importadas as bibliotecas necessárias para o funcionamento do crawler.

### `os`
A biblioteca `os` é usada para trabalhar com diretórios e caminhos de arquivos. Neste exercício, ela será utilizada para verificar se a pasta onde os HTMLs serão salvos já existe e, se necessário, criá-la.

### `requests`
A biblioteca `requests` é usada para fazer requisições HTTP. Ela permite acessar o conteúdo de uma página web a partir de sua URL.

### `BeautifulSoup`
A classe `BeautifulSoup`, da biblioteca `bs4`, é usada para interpretar o HTML baixado e transformá-lo em uma estrutura em árvore. Isso facilita muito a busca e a navegação pelos elementos da página.

### `urljoin`
A função `urljoin`, do módulo `urllib.parse`, serve para transformar links relativos em links completos. Isso é importante porque muitos links no HTML da Wikipédia aparecem apenas como caminhos, por exemplo:

`/wiki/Probability`

e precisam ser convertidos para:

`https://en.wikipedia.org/wiki/Probability`

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

## B. Definição das variáveis principais

Aqui são definidas três variáveis importantes para o funcionamento do programa.

### `pasta_destino`
Essa variável define onde os arquivos HTML serão armazenados dentro do ambiente do Colab.

No caso deste trabalho, os arquivos serão salvos em:

`/content/html_pages`

### `url_inicial`
Essa variável guarda a URL da página de partida do crawler. O artigo escolhido foi a página da Wikipédia sobre Thomas Bayes.

### `headers`
Essa variável armazena um dicionário com o cabeçalho HTTP da requisição.

Foi incluído o campo `User-Agent` para identificar o cliente que está fazendo a requisição. Em muitos sites, isso ajuda a evitar recusas simples do servidor.

```python
pasta_destino = "/content/html_pages"
url_inicial = "https://en.wikipedia.org/wiki/Thomas_Bayes"
headers = {"User-Agent": "aula-cpa"}
```

## C. Criação da pasta de saída

Antes de salvar qualquer arquivo, o programa precisa garantir que a pasta de destino exista.

### `os.path.exists(pasta_destino)`
Verifica se a pasta já existe.

### `os.makedirs(pasta_destino)`
Caso a pasta ainda não exista, ela é criada automaticamente.

Essa etapa evita erro ao tentar gravar arquivos em um diretório inexistente.

```python
if not os.path.exists(pasta_destino):
    os.makedirs(pasta_destino)
```

## D. Função para baixar e salvar uma página HTML

Nesta etapa, foi criada a função `baixar_html`, responsável por acessar uma URL, salvar seu conteúdo HTML em arquivo e retornar esse conteúdo em formato de texto.

### Parâmetros da função

A função recebe dois parâmetros:

- `url`: endereço da página que será acessada;
- `nome_arquivo`: nome com que o HTML será salvo.

### Etapas executadas pela função

#### 1. Fazer a requisição
A função usa `requests.get(url, headers=headers)` para acessar a página desejada.

#### 2. Verificar o código de resposta
Se o servidor responder com código `200`, significa que a requisição foi bem-sucedida.

#### 3. Montar o caminho do arquivo
O caminho completo do arquivo é montado com `os.path.join()`.

#### 4. Salvar o HTML
O conteúdo da página é salvo em um arquivo `.html` usando `open(..., "w", encoding="utf-8")`.

#### 5. Retornar o conteúdo
Ao final, a função retorna o HTML da página em formato de texto, o que permite que esse conteúdo seja usado imediatamente no restante do programa.

### Caso a página não seja baixada
Se o código de resposta não for `200`, a função retorna `None`.

```python
def baixar_html(url, nome_arquivo):
    retorno = requests.get(url,headers=headers)

    if retorno.status_code == 200:
        caminho = os.path.join(pasta_destino, nome_arquivo)

        with open(caminho, "w", encoding="utf-8") as f:
            f.write(retorno.text)

        return retorno.text
    else:
        print(f"Erro ao acessar {url} - Status: {retorno.status_code}")
        return None
```

## E. Download da página inicial

Nesta etapa, o programa chama a função `baixar_html()` para baixar a página inicial do crawler.

A URL utilizada é a do artigo sobre Thomas Bayes, e o arquivo será salvo com o nome:

`Thomas_Bayes.html`

Além de salvar o arquivo na pasta definida, a função também retorna o HTML da página, que é armazenado na variável `html`.

Essa variável será usada em seguida para criar o objeto `soup`.

```python
html = baixar_html(url_inicial, "Thomas_Bayes.html")
```

## F. Criação do objeto Beautiful Soup

Se o HTML da página inicial foi baixado corretamente, o programa cria um objeto `BeautifulSoup`.

### `BeautifulSoup(html, "html.parser")`
Essa instrução recebe o HTML bruto e o transforma em uma estrutura organizada, semelhante a uma árvore DOM.

Isso permite:

- acessar elementos da página;
- buscar tags específicas;
- extrair atributos;
- navegar pelo documento HTML.

### Lista `links`
Também é criada uma lista vazia chamada `links`, que será usada para armazenar os links internos encontrados na página, sem repetição de URLs.

```python
if html:
    soup = BeautifulSoup(html, "html.parser")
    links = []
```

## G. Busca dos links internos da página

Nesta etapa, o programa percorre os links do HTML da página inicial.

O método utilizado é `find_all("a", href=True)`: comando que procura todas as tags `<a>` com o atributo `href`.

A tag `<a>` representa links em HTML.

Para cada tag encontrada, o programa extrai o valor do atributo `href`, que indica o destino do link.

Em seguida, é aplicado o filtro `href.startswith("/wiki/")`. Serão selecionados apenas os links internos da Wikipédia, que possuem o padrão.

Quando um link atende a esse critério, ele ainda está em formato relativo. Por isso, o programa usa `urljoin("https://en.wikipedia.org", href)` para transformar o caminho relativo em uma URL completa.

Depois disso, é feita uma verificação para evitar links duplicados. Se a URL completa ainda não estiver na lista `links`, ela é adicionada.

O programa imprime para controle a url completa e, ao final, a quantidade de links internos encontrados.

```python
    for tag in soup.find_all("a", href=True):
        href = tag["href"]

        if href.startswith("/wiki/"):
            url_completa = urljoin("https://en.wikipedia.org", href)

            if url_completa not in links:
                print(url_completa)
                links.append(url_completa)

    print("Quantidade de links internos encontrados:", len(links))
```

## H. Download das páginas internas encontradas

Depois de montar a lista `links`, o programa passa para a etapa de download das páginas internas encontradas na página inicial.

Nesta parte, é feito um laço `for` para percorrer cada URL armazenada na lista. Para cada link, o programa cria um nome de arquivo a partir da última parte do endereço.

Isso é feito com:

`url.split("/")[-1] + ".html"`

A lógica é a seguinte:

- `split("/")` separa a URL em partes usando a barra `/`;
- `[-1]` pega a última parte da URL;
- `+ ".html"` adiciona a extensão do arquivo.

Por exemplo:

- URL: `https://en.wikipedia.org/wiki/Probability`
- nome gerado: `Probability.html`

Depois de gerar o nome do arquivo, o programa chama novamente a função `baixar_html(url, nome_arquivo)`. Essa função acessa a URL, baixa o conteúdo HTML da página e salva o arquivo na pasta definida anteriormente.

Dessa forma, cada link interno encontrado na página inicial é transformado em um arquivo HTML salvo no ambiente do Colab.

```python
    for url in links:
        nome_arquivo = url.split("/")[-1] + ".html"
        baixar_html(url, nome_arquivo)
```
